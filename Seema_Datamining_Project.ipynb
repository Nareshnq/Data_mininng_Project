```python
import numpy as np  

df1 = df.copy() 

# Convert infinite values to NaN
df1.replace([np.inf, -np.inf], np.nan, inplace=True)

# List of numerical variables
numeric_columns = ['est_diameter_min', 'est_diameter_max', 'relative_velocity', 'miss_distance', 'absolute_magnitude']

def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Detect and print outliers for each numerical variable
for col in numeric_columns:
    outliers = detect_outliers(df1[col])
    print(f"Outliers in {col} (IQR method):\n", outliers, "\n")
```

##  <font color='blue'>8.1 Chi-square test to assess independence between two categorical variables

```python
import scipy.stats as stats

# Create a contingency table
contingency_table = pd.crosstab(df['name'], df['hazardous'])
print("Contingency Table:")
print(contingency_table)

# Perform the Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nThere is a significant association between 'name' and 'hazardous' (reject null hypothesis).")
else:
    print("\nThere is no significant association between 'name' and 'hazardous' (fail to reject null hypothesis).")
```

## <font color='blue'>8.3 Mann-Whitney U test to compare means between two groups.

#### Let us consider the hazardous column

```python
from scipy.stats import mannwhitneyu

variables = ['absolute_magnitude']


# Split data into two groups based on 'hazardous'
group1 = df[df['hazardous'] == True]
group2 = df[df['hazardous'] == False]

# Perform Mann-Whitney U test for each variable
results = {}
for var in variables:
    stat, p_value = mannwhitneyu(group1[var], group2[var])
    results[var] = {'U statistic': stat, 'p-value': p_value}

# Display results
for var, result in results.items():
    print(f"{var}: U statistic = {result['U statistic']}, p-value = {result['p-value']}")
    alpha = 0.05
    if result['p-value'] < alpha:
        print(f"Reject the null hypothesis. There is a significant difference in {var} between hazardous and non-hazardous objects.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant difference in {var} between hazardous and non-hazardous objects.")
    print()
    
plt.figure(figsize=(10, 6))
sns.boxplot(x='hazardous', y='absolute_magnitude', data=df)
plt.title('Box plot of Absolute Magnitude by Hazardous Status')
plt.xlabel('Hazardous')
plt.ylabel('Absolute Magnitude')
plt.show()
```

## <font color='blue'>8.4 Analysis of Variance (ANOVA) to compare means across multiple groups

```python
from scipy.stats import f_oneway

# Read the dataset
file_path = "neo.csv"
df = pd.read_csv(file_path)

# Prepare the data by grouping prices by city
grouped_data = df.groupby('miss_distance')['absolute_magnitude'].apply(list)

# Perform ANOVA
anova_result = f_oneway(*grouped_data)
print(f"F-statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")

# Interpretation
alpha = 0.05
if anova_result.pvalue < alpha:
    print("There is a statistically significant difference in miss_distance and absolute magnitude.")
else:
    print("There is no statistically significant difference in miss_distance and absolute magnitude.")

```

## <font color='blue'>9 .Check for presence of collinearity and multi-collinearity 


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# List of independent variables (excluding the dependent variable 'price')
ivar = ['est_diameter_min', 'est_diameter_max', 'relative_velocity', 'miss_distance', 'absolute_magnitude']

# Drop rows with missing values to calculate VIF
df = df[ivar].dropna()

# Correlation Matrix
plt.figure(figsize=(8, 6))
corr_matrix = df[ivar].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1, annot_kws={"size": 8})
plt.title('Correlation Matrix')
plt.show()
```

##### <font color='red'>Correlation Matrix Heatmap:
The heatmap displays the correlation coefficients between the independent variables (est_diameter_min, est_diameter_max, relative_velocity, miss_distance, and absolute_magnitude).
Each cell in the matrix represents the correlation coefficient between two variables.
The color scale indicates the strength and direction of the correlation:
Dark red indicates positive correlation (values close to 1), meaning that when one variable increases, the other tends to increase as well.
Dark blue indicates negative correlation (values close to -1), implying that when one variable increases, the other tends to decrease.
White cells represent no significant correlation (values close to 0).
Observations:
The diagonal from top left to bottom right consists of 1s because it represents each variables correlation with itself (which is always perfect).
Notable observations:
est_diameter_min and est_diameter_max have a very high positive correlation (0.99), which makes sense since they are related measures of size.
relative_velocity shows moderate positive correlations with both est_diameter_min (0.22) and est_diameter_max (0.24), suggesting that larger objects tend to move faster relative to smaller ones.
miss_distance has a weak positive correlation with relative_velocity (0.14), indicating that closer objects tend to move faster.
absolute_magnitude has a moderate negative correlation with relative_velocity (-0.25), implying that brighter objects move more slowly.
High correlations between predictors (est_diameter_min and est_diameter_max) can lead to multicollinearity, affecting coefficient estimates and interpretability.


```python
# Calculate VIF
def calculate_vif(data, variables):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = variables
    vif_data["VIF"] = [variance_inflation_factor(data[variables].values, i) for i in range(len(variables))]
    vif_data = vif_data.sort_values(by='VIF', ascending=False)  # Sort by VIF values in descending order
    return vif_data

# Calculate VIF for the independent variables
vif_data = calculate_vif(df, ivar)
print("Variance Inflation Factor (VIF) in descending order:")
print(vif_data)
```

The variables with the highest VIFs are est_diameter_min and est_diameter_max (which are infinite). These variables likely have strong multicollinearity.
Consider removing one of these variables to improve the models stability and interpretability.

```python
# Drop 'est_diameter_min' from the list of independent variables
ivar.remove('est_diameter_min')

# Calculate VIF for the modified independent variables
vif_data = calculate_vif(df, ivar)
print("Variance Inflation Factor (VIF) after dropping 'est_diameter_min':")
print(vif_data)
```

## <font color='blue'>For regression show: R2, Adjusted R2, RMSE, correlation matrix, and p-values of independent variables 

```python
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Assuming you have already split your data into X_train, X_test, y_train, y_test

# Step 1: Fit the regression model
model = sm.OLS(y_train, sm.add_constant(X_train))
results = model.fit()

# Step 2: Predict on the test set
y_pred = results.predict(sm.add_constant(X_test))

# Step 3: Calculate R-squared and adjusted R-squared
r_squared = r2_score(y_test, y_pred)
num_obs = len(y_test)
num_features = X_test.shape[1]
adjusted_r_squared = 1 - (1 - r_squared) * ((num_obs - 1) / (num_obs - num_features - 1))

# Step 4: Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Step 5: Correlation Matrix
correlation_matrix = np.corrcoef(X_train, rowvar=False)

# Step 6: P-values of independent variables
p_values = results.pvalues

# Print the results
print("R-squared:", r_squared)
print("Adjusted R-squared:", adjusted_r_squared)
print("RMSE:", rmse)
print("Correlation Matrix:\n", correlation_matrix)
print("P-values of independent variables:\n", p_values)
```

##### <font color='red'>R-squared and Adjusted R-squared:

R-squared measures the proportion of the variance in the dependent variable (hazardous) that is predictable from the independent variables (est_diameter_min, est_diameter_max, relative_velocity, miss_distance) in the model. In this case, R-squared is 0.33, indicating that approximately 33% of the variance in hazardous can be explained by the independent variables.
Adjusted R-squared adjusts the R-squared value for the number of predictors in the model and the sample size. It penalizes for adding unnecessary predictors. The adjusted R-squared is slightly lower than the R-squared, suggesting that the model may not be significantly improved by additional predictors.
RMSE (Root Mean Squared Error):

RMSE measures the average magnitude of the errors between predicted and observed values. In this case, RMSE is 2.37, which means, on average, the model's predictions are off by approximately 2.37 units of hazardous.

##### <font color='red'>Correlation Matrix:

The correlation matrix shows the pairwise correlations between the independent variables. Each cell represents the correlation coefficient between two variables. For example, the correlation coefficient between est_diameter_min and est_diameter_max is approximately 1, indicating a strong positive correlation between these two variables.
##### <font color='red'>P-values of Independent Variables:

The p-values associated with each independent variable test the null hypothesis that the coefficient of the variable is equal to zero (i.e., the variable has no effect on the dependent variable). In this case, all p-values are very close to zero, suggesting that all independent variables are statistically significant in predicting the hazardous outcome.
Overall, the model seems to have some predictive power, as indicated by the statistically significant p-values and the non-zero R-squared value.
