```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a sample DataFrame with duplicate values
np.random.seed(0)
data = {'value': np.random.randint(0, 10, size=100)}
df = pd.DataFrame(data)

# Concatenate the DataFrame to introduce duplicate values
df = pd.concat([df, df], ignore_index=True)

# Plotting histogram with duplicates
plt.figure(figsize=(8, 6))
plt.hist(df['value'], bins=range(11), edgecolor='black')
plt.title('Histogram with Duplicate Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xticks(range(10))
plt.grid(axis='y', alpha=0.75)
plt.show()

```

```python
# Remove duplicates
df_unique = df.drop_duplicates()

# Plotting histogram without duplicates
plt.figure(figsize=(8, 6))
plt.hist(df_unique['value'], bins=range(11), edgecolor='black')
plt.title('Histogram without Duplicate Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xticks(range(10))
plt.grid(axis='y', alpha=0.75)
plt.show()

```
